# Q1. Data Visualization and Statistical Measures:
For this question, you are required to analyse the iris dataset (iris.csv) using Python. Perform all
possible data visualization techniques (histograms, scatter plots, box plots, etc.) on all numerical
columns of the dataset. Additionally, calculate all possible statistical measures (mean, median,
mode, standard deviation, etc.) for each numerical column. 

```python
#Q.1
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df=pd.read_csv("C:/Users/hp/Downloads/Iris.csv")
iris_df
```

```python
print(iris_df.head())
```

```python
print("Summary Statistics:")
print(iris_df.describe())
```

```python
iris_df.hist(figsize=(10,8))
plt.suptitle('Histograms of Numerical columns')
plt.show()
```

```python
iris_df.boxplot(figsize=(10,6))
plt.title('Box Plot of Numerical Columns')
plt.show()
```

```python
sns.pairplot(iris_df, hue='Species')
plt.suptitle('Scatter Plot Matrix')
plt.show()
```

```python
print(iris_df["SepalLengthCm"].mean())
print(iris_df["SepalWidthCm"].mean())
print(iris_df["PetalLengthCm"].mean())
print(iris_df["PetalWidthCm"].mean())
```

```python
print(iris_df["SepalLengthCm"].mode())
print(iris_df["SepalWidthCm"].mode())
print(iris_df["PetalLengthCm"].mode())
print(iris_df["PetalWidthCm"].mode())
```

```python
print(iris_df["SepalLengthCm"].median())
print(iris_df["SepalWidthCm"].median())
print(iris_df["PetalLengthCm"].median())
print(iris_df["PetalWidthCm"].median())
```
